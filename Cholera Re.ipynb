# Cholera Research

## Cholera Data Cleaning

```python
# Import any necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# importing the data
cholera = pd.read_csv("cholera.csv")
```

```python
#viewing the dataset
cholera.head()
```

```python
# The total number of reported cases in Africa
total_cases_in_africa = cholera[cholera['Continent'] == 'Africa']['Reported cases'].sum()
print("Total reported cases in Africa:", total_cases_in_africa)
```

```python
# The total number of reported deaths in Africa
total_deaths_in_africa = cholera[cholera['Continent'] == 'Africa']['Reported deaths'].sum()
print("Total reported deaths in Africa:", total_deaths_in_africa)
```

```python
# Top 20 cholera case in Africa
african_countries_1995_to_2016 = cholera[(cholera['Continent'] == 'Africa') & (cholera['Year'] >= 1995) & (cholera['Year'] <= 2016)]
sum_reported_cases_by_country = african_countries_1995_to_2016.groupby('Country')['Reported cases'].sum().reset_index()
top_20_african_countries_1995_to_2016 = sum_reported_cases_by_country.sort_values(by='Reported cases', ascending=False).head(20)
print("Top 20 African countries with the highest sum of reported cases from 1995 to 2016:")
print(top_20_african_countries_1995_to_2016)
```

```python
# Top 20 cholera Deaths in Africa
african_countries_1995_to_2016 = cholera[(cholera['Continent'] == 'Africa') & (cholera['Year'] >= 1995) & (cholera['Year'] <= 2016)]
sum_reported_deaths_by_country = african_countries_1995_to_2016.groupby('Country')['Reported deaths'].sum().reset_index()
top_20_african_countries_deaths_1995_to_2016 = sum_reported_deaths_by_country.sort_values(by='Reported deaths', ascending=False).head(20)
print("Top 20 African countries with the highest sum of reported deaths from 1995 to 2016:")
print(top_20_african_countries_deaths_1995_to_2016)
```

```python
# Top 10 cholera case in Africa
african_countries_1995_to_2016 = cholera[(cholera['Continent'] == 'Africa') & (cholera['Year'] >= 1995) & (cholera['Year'] <= 2016)]
sum_reported_cases_by_country = african_countries_1995_to_2016.groupby('Country')['Reported cases'].sum().reset_index()
top_10_african_countries_cases_1995_to_2016 = sum_reported_cases_by_country.sort_values(by='Reported cases', ascending=False).head(10)
print("Top 10 African countries with the highest sum of reported cases from 1995 to 2016:")
print(top_10_african_countries_cases_1995_to_2016)
```

```python
# Top 10 cholera Deaths in Africa
african_countries_1995_to_2016 = cholera[(cholera['Continent'] == 'Africa') & (cholera['Year'] >= 1995) & (cholera['Year'] <= 2016)]
sum_reported_deaths_by_country = african_countries_1995_to_2016.groupby('Country')['Reported deaths'].sum().reset_index()
top_10_african_countries_deaths_1995_to_2016 = sum_reported_deaths_by_country.sort_values(by='Reported deaths', ascending=False).head(10)
print("Top 10 African countries with the highest sum of reported deaths from 1995 to 2016:")
print(top_10_african_countries_deaths_1995_to_2016)
```

```python
# Top 10 cholera case in Africa
import matplotlib.pyplot as plt
african_countries_1995_to_2016 = cholera[(cholera['Continent'] == 'Africa') & (cholera['Year'] >= 1995) & (cholera['Year'] <= 2016)]
sum_reported_cases_by_country = african_countries_1995_to_2016.groupby('Country')['Reported cases'].sum().reset_index()
top_10_african_countries_cases_1995_to_2016 = sum_reported_cases_by_country.sort_values(by='Reported cases', ascending=False).head(10)
plt.figure(figsize=(12, 6))
plt.bar(top_10_african_countries_cases_1995_to_2016['Country'], top_10_african_countries_cases_1995_to_2016['Reported cases'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Sum of Reported Cases')
plt.title('Top 10 African Countries with Highest Reported Cases (1995-2016)')
plt.xticks(rotation=45, ha='right')
for i, val in enumerate(top_10_african_countries_cases_1995_to_2016['Reported cases']):
    plt.text(i, val, str(val), ha='center', va='bottom')
plt.tight_layout()
plt.show()
```

```python
# Top 10 cholera Deaths in Africa
import matplotlib.pyplot as plt
african_countries_1995_to_2016 = cholera[(cholera['Continent'] == 'Africa') & (cholera['Year'] >= 1995) & (cholera['Year'] <= 2016)]
sum_reported_deaths_by_country = african_countries_1995_to_2016.groupby('Country')['Reported deaths'].sum().reset_index()
top_10_african_countries_deaths_1995_to_2016 = sum_reported_deaths_by_country.sort_values(by='Reported deaths', ascending=False).head(10)
plt.figure(figsize=(12, 6))
plt.bar(top_10_african_countries_deaths_1995_to_2016['Country'], top_10_african_countries_deaths_1995_to_2016['Reported deaths'], color='salmon')
plt.xlabel('Country')
plt.ylabel('Sum of Reported Deaths')
plt.title('Top 10 African Countries with Highest Reported Deaths (1995-2016)')
plt.xticks(rotation=45, ha='right')
for i, val in enumerate(top_10_african_countries_deaths_1995_to_2016['Reported deaths']):
    plt.text(i, val, str(val), ha='center', va='bottom')
plt.tight_layout()
plt.show()
```

```python
# Top 10 cholera case and  Deaths in Africa
african_countries_1995_to_2016 = cholera[(cholera['Continent'] == 'Africa') & (cholera['Year'] >= 1995) & (cholera['Year'] <= 2016)]
sum_reported_cases_deaths_by_country = african_countries_1995_to_2016.groupby('Country')[['Reported cases', 'Reported deaths']].sum().reset_index()
top_10_african_countries_cases_deaths_1995_to_2016 = sum_reported_cases_deaths_by_country.sort_values(by='Reported cases', ascending=False).head(10)
plt.figure(figsize=(12, 6))
bar_width = 0.4
index = range(len(top_10_african_countries_cases_deaths_1995_to_2016))
plt.bar(index, top_10_african_countries_cases_deaths_1995_to_2016['Reported cases'], width=bar_width, color='skyblue', label='Reported Cases')
plt.bar([i + bar_width for i in index], top_10_african_countries_cases_deaths_1995_to_2016['Reported deaths'], width=bar_width, color='salmon', label='Reported Deaths')
plt.xlabel('Country')
plt.ylabel('Sum of Reported Cases and Deaths')
plt.title('Top 10 African Countries with Highest Reported Cases and Deaths (1995-2016)')
plt.xticks([i + bar_width / 2 for i in index], top_10_african_countries_cases_deaths_1995_to_2016['Country'], rotation=45, ha='right')
for i, (cases, deaths) in enumerate(zip(top_10_african_countries_cases_deaths_1995_to_2016['Reported cases'], top_10_african_countries_cases_deaths_1995_to_2016['Reported deaths'])):
    plt.text(i, cases + 500, f'Cases: {cases}\nDeaths: {deaths}', ha='center', va='bottom')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Trend in Reported Case in Africa
african_countries_1995_to_2016 = cholera[(cholera['Continent'] == 'Africa') & (cholera['Year'] >= 1995) & (cholera['Year'] <= 2016)]
sum_reported_cases_by_year = african_countries_1995_to_2016.groupby('Year')['Reported cases'].sum().reset_index()
sorted_data = sum_reported_cases_by_year.sort_values(by='Year')
plt.figure(figsize=(12, 8))
plt.plot(sorted_data['Year'], sorted_data['Reported cases'], marker='o', label='Reported Cases')
for i, row in sorted_data.iterrows():
    plt.annotate(f"{row['Reported cases']}", (row['Year'], row['Reported cases']), textcoords="offset points", xytext=(0, 10), ha='center')
plt.title('Trend in Reported Cases in Africa (1995-2016)')
plt.xlabel('Year')
plt.ylabel('Reported Cases')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
#  Trend in Reported Deaths in Africa
african_countries_1995_to_2016 = cholera[(cholera['Continent'] == 'Africa') & (cholera['Year'] >= 1995) & (cholera['Year'] <= 2016)]
sum_reported_deaths_by_year = african_countries_1995_to_2016.groupby('Year')['Reported deaths'].sum().reset_index()
sorted_data = sum_reported_deaths_by_year.sort_values(by='Year')
plt.figure(figsize=(12, 8))
plt.plot(sorted_data['Year'], sorted_data['Reported deaths'], marker='o', label='Reported Deaths')
for i, row in sorted_data.iterrows():
    plt.annotate(f"{row['Reported deaths']}", (row['Year'], row['Reported deaths']), textcoords="offset points", xytext=(0, 10), ha='center')
plt.title('Trend in Reported Deaths in Africa (1995-2016)')
plt.xlabel('Year')
plt.ylabel('Reported Deaths')
plt.legend()
plt.grid()
# Show the chart
plt.tight_layout()
plt.show()
```
